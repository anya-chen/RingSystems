```python
def get_df(directory,subset):
    csv_files = glob.glob(os.path.join(directory, subset+"*.csv"))
    frames = []
    for f in csv_files:
        df = pd.read_csv(f)
        frames.append(df)

    results_df = pd.concat(frames)
    return results_df
```

```python
plantsDF = get_df(ringDir,'plants')
plantsDF
```

```python
plantsDF_compound_noRing = get_df(compound_noRingDir,'plants')
plantsDF_compound_noRing
```

```python
nPlants_noStereo = len(set(plantsDF.MoleculeConID)) + len(set(plantsDF_compound_noRing.conID))
nPlants_noStereo
```

```python
def get_idSet(df):
    idSet = set()
    for i in list(df.mol_stereoIDs):
        if '[' in i:
            ids = i.replace('[','').replace(']','').replace("'","").replace(' ','')
            id_list = ids.split(',')
            for item in id_list:
                idSet.add(item)
        else:
            idSet.add(i)
    return idSet
```

```python
plants_idSet= get_idSet(plantsDF)
plants_idSet_noRing = get_idSet(plantsDF_compound_noRing)
nPlants_Stereo = len(plants_idSet) + len(plants_idSet_noRing)
nPlants_Stereo
```

### number of macrocycles

```python
from rdkit.Chem import MolFromSmarts

def macrocycle(mol):
    if mol.HasSubstructMatch(MolFromSmarts('[r;!r3;!r4;!r5;!r6;!r7;!r8;!r9;!r10;!r11]')):
        return 1
    else:
        return 0
```
