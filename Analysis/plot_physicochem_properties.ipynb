```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import matplotlib.font_manager
plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = 'sans-serif'
```

```python
NPringDF = pd.read_csv('/home/ychen/projects/nps_ringsys/20210901_analysis/NP_all_descriptors.csv', sep=';')
ZINCringDF = pd.read_csv('/home/ychen/projects/nps_ringsys/20210901_analysis/ZINC_all_descriptors.csv', sep=';')
```

```python
labelDict = {'N': 'Number of nitrogen atoms',
                   'O': 'Number of oxygen atoms',
                   'chiral': 'Number of chiral centers',
                   'MW': 'Molecular weight [DA]',
                   'heavy_atoms': 'Number of heavy atoms',
                   'h_acc': 'Number of hydrogen-bond acceptors',
                   'h_don': 'Number of hydrogen-bond donors',
                   'logP': 'logP (o/w)',
                   'TPSA': 'Topological polar surface area',
                   'numAro': 'Number of aromatic atoms',
                   'formalCharge': 'Formal Charge',
                   'numRings': 'Number of rings',
                   'bridgeheadAtoms': 'Number of bridgehead atoms',
                   'frac_csp3': 'Fraction of Csp3 atoms'}
```

```python
propertyColumn = 'formalCharge'
figure, (ax,ax2) = plt.subplots(2,1,sharex=True, figsize=(10.5, 7.5))
NP_heights, NP_bins = np.histogram(NPringDF[propertyColumn], bins=np.arange(NPringDF[propertyColumn].max() + 2))
# print(NP_bins)
ZINC_heights, ZINC_bins = np.histogram(ZINCringDF[propertyColumn], bins=np.arange(ZINCringDF[propertyColumn].max() + 2))

NP_heights_fraction = NP_heights/len(NPringDF)
ZINC_heights_fraction = ZINC_heights/len(ZINCringDF)

width = 0.35

if NP_bins[-1] > ZINC_bins[-1]:
    max_range = NP_bins[-1]
else:
    max_range = ZINC_bins[-1]

# set x-axis here and adapt it if needed
x_axis = [-2,-1, 0, 1,2, 3,4,5,6]

ax.bar(NP_bins[:-1]-width/2, NP_heights_fraction, width=width, label='NPs', color='#145c37')
ax.bar(ZINC_bins[:-1]+width/2, ZINC_heights_fraction, width=width, label='SCs', color='#ffa600')
ax2.bar(NP_bins[:-1]-width/2, NP_heights_fraction, width=width, label='NPs', color='#145c37')
ax2.bar(ZINC_bins[:-1]+width/2, ZINC_heights_fraction, width=width, label='SCs', color='#ffa600')

# zoom-in / limit the view to different portions of the data
ax.set_ylim(0.925,1)
ax2.set_ylim(0,0.065)

# hide the spines between ax and ax2
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

d = .01  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal


ax.tick_params(labelsize=24)
ax2.tick_params(labelsize=24)
ax.set_xticks(x_axis)
ax2.set_xticks(x_axis)
ax.set_xlim(left=-2, right=6.5)  # set the limit of the x-axis here
ax2.set_xlabel(labelDict[propertyColumn], fontsize=30)
ax.legend(fontsize=24)

#plt.ylabel('Fraction', fontsize=30)
figure.text(.006, .5, 'Fraction', fontsize=30, ha='center', va='center', rotation='vertical')

plt.savefig('/data/local/ringsys/202112_figures/formalCharge.png')
plt.show()
```
