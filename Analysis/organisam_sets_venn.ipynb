```python
import venn
# https://github.com/tctianchi/pyvenn
```

```python
uniqueRingFolder = '/home/ychen/projects/nps_ringsys/20210618_preprocessing/organismSets/uniqueRingSystems/'
```

```python
plantsRings_noStereo = pd.read_csv(uniqueRingFolder+'plants_uniqueRingSystems_noStereo.txt',sep='\t')
bacteriaRings_noStereo = pd.read_csv(uniqueRingFolder+'bacteria_uniqueRingSystems_noStereo.txt',sep='\t')
fungiRings_noStereo = pd.read_csv(uniqueRingFolder+'fungi_uniqueRingSystems_noStereo.txt',sep='\t')
marineRings_noStereo = pd.read_csv(uniqueRingFolder+'marine_uniqueRingSystems_noStereo.txt',sep='\t')
```

```python
set1 = set(plantsRings_noStereo.ringSmiles_noStereo)
set2 = set(bacteriaRings_noStereo.ringSmiles_noStereo)
set3 = set(fungiRings_noStereo.ringSmiles_noStereo)
set4 = set(marineRings_noStereo.ringSmiles_noStereo)
```

```python
color_lists = [
    #R G B A
    [138, 184, 99, 0.5],
    [161, 123, 185, 0.5],
    [128,128,128, 0.6],
    [0, 105, 148, 0.4],
]

set_colors = [
    [i[0] / 255.0, i[1] / 255.0, i[2] / 255.0, i[3]]
    for i in color_lists
]
```

```python
labels = venn.get_labels([set1,set2,set3,set4], fill=['number'])
fig, ax = venn.venn4(labels, names=['plants', 'bacteria', 'fungi', 'marine'],dpi=300,colors=set_colors)
```
